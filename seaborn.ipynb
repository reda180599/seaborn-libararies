```python
#seaborn library   for visualization


# real application on seabon libarariy

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt


import seaborn as sns







```

```python
#read data from .csv file


df = pd.read_csv('data/Salaries.csv')

df
```

```python

# correlation  -- relation between the columns

# correlation resulut may be positive or negative  or non


d = df.corr()
print(d)
```

```python


#heat map   description for correlation 

sns.heatmap(d, annot=True)


plt.show()
```

```python
sns.pairplot(d)
```
